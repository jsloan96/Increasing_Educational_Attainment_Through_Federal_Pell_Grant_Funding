This notebook merges some additional features of the IPEDS data set with the Pell Partnership data set.

```python
import pandas as pd
```

```python
IPEDS = pd.read_csv('UTF8_delta_public_00_12.csv', header=0)
IPEDS.head()
```

Select the features from the IPEDS data set that we what to merge.

```python
IPEDS_sel = IPEDS[['unitid', 'academicyear','admssn', 'applcn']]
IPEDS_sel
```

```python
ThePellPartnership = pd.read_csv('ThePellPartnership_EdTrust.csv', header=0)
ThePellPartnership.head()
```

Rename the Pell Partnership ID column to match IPEDS ID column name.

```python
ThePellPartnership = ThePellPartnership.rename(columns={"IPEDS Unit ID": "unitid"})
```

```python
ThePellPartnership.info()
```

Select academic year 2012 from the IPEDS data set because the Pell Partnership data set only consists of data from the 2012 acedemic school year.

```python
# Select only the rows where 'academicyear' == 2012 in IPEDS_sel dataframe
IPEDS_sel_2012 = IPEDS_sel[IPEDS_sel['academicyear'] == 2012]

# Merge 'admssn' from IPEDS_sel_2012 with ThePellPartnership using 'unitid' as key
merged_df = pd.merge(IPEDS_sel_2012[['unitid','admssn', 'applcn']], ThePellPartnership, on='unitid', how='inner')

# Print the resulting dataframe
merged_df.info()
```

Make sure there are no duplicate ID's after the merge.

```python
# Drop duplicates based on 'unitid'
merged_df = merged_df.drop_duplicates(subset='unitid')

# Print the resulting dataframe
merged_df.info()
```

```python
merged_df
```

```python
merged_df = merged_df.dropna()
merged_df
```

## Create a new variable 'addmission rate' using existing features

acceptance rate = total number of admitted students / total number of applicants

```python
#merged_df = merged_df.loc[:, ['unitid', 'admssn', 'applcn']]
merged_df['admssn_rate'] = round((merged_df['admssn'] / merged_df['applcn']) * 100, 1)
merged_df['Number of Undergraduates Recieving Pell Grants'] = round(merged_df['Number of Undergraduates'] * (merged_df['Percent Pell Recipients Among Undergraduates'] / 100), 0)
merged_df['avg_pell_grant_per_student'] = round(merged_df['Total Pell Dollars Received, 2012-13'] / merged_df['Number of Undergraduates Recieving Pell Grants'], 2)

#merged_df['admssn_rate'] = (merged_df['admssn'] / merged_df['applcn']) * 100
merged_df
```
